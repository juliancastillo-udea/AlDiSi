# **Montecarlo Integrales**
## Estimar integrales usando Montecarlo

La simulación mediante métodos MonteCarlo ha sido muy utilizada en el cálculo de integrales, supongamos $f(x)$ una función en la cual deseamos calcular su integral definida entre $0$ y $1$.

$\theta = \int_{a}^{b} \ f(x) \ dx$


Para calcular el valor de $\theta$, tenemos a $U$ distribuida uniformemente entre $(0,1)$, entonces podemos expresar a $\theta$ como:

$\theta = E[f(U)]$

Si $U_1$, $U_2$, ..., $U_k$ son variables aleatorias independientes y uniformes en $(0,1)$, esto implica que $f(U_1)$, $f(U_2)$, ..., $f(U_k)$ son variables aleatorias independientes e idénticamente distribuidas con media $\theta$. Por la ley fuerte de los grandes números, se tiene, con probabilidad 1,

$\sum_{i=1}^{n} \frac{f(U_1)}{k}\rightarrow E[f(U)]=\theta \ \ cuando \ k\rightarrow \infty$


Por lo tanto se puede estimar $\theta$ generando un gran número de números aleatorios $u_i$ y considerando nuestra aproximación a $\theta$ el valor promedio de $f(u)$


## Ley de los grandes números

*   En la teoría de la probabilidad, bajo el término genérico de ley de los grandes números se engloban varios teoremas que describen el comportamiento del promedio de una sucesión de variables aleatorias conforme aumenta su número de ensayos.
*   Estos teoremas prescriben condiciones suficientes para garantizar que dicho promedio converja al promedio de las esperanzas de las variables aleatorias involucradas. Las distintas formulaciones de la ley de los grandes números (y sus condiciones asociadas) especifican la convergencia de formas distintas.
*   Las leyes de los grandes números explican por qué el promedio de una muestra al azar de una población de gran tamaño tenderá a estar cerca de la media de la población completa.
*   Cuando las variables aleatorias tienen una varianza finita, el teorema del límite central extiende nuestro entendimiento de la convergencia de su promedio describiendo la distribución de diferencias estandarizadas entre la suma de variables aleatorias y el valor esperado de esta suma: sin importar la distribución subyacente de las variables aleatorias, esta diferencia estandarizada converge a una variable aleatoria normal estándar.
*   La frase "ley de los grandes números" es también usada ocasionalmente para referirse al principio de que la probabilidad de que cualquier evento posible (incluso uno improbable) ocurra al menos una vez en una serie, aumenta con el número de eventos en la serie. Por ejemplo, la probabilidad de que un individuo gane la lotería es bastante baja; sin embargo, la probabilidad de que alguien gane la lotería es bastante alta, suponiendo que suficientes personas comprasen boletos de lotería.

## `Ejercicio 1`

Encuentre una aproximación usando el metodo de MonteCarlo para el área delimitada por el Seno y Coseno, estos valores aplican entre 0 y 1 tanto para  x  como para  y. Realice un algoritmo que calcule el área. Aplica solo para los valores entre curvas.

```python
replicas = 10000 #@param {type:"slider", min:100, max:10000, step:100}
```

```python
hx = []
hy = []
tx = []
ty = []
for i in range(0,replicas+1, 1):
    valor = i/replicas
    a = round(mt.sin(valor),4)
    b = round(mt.cos(valor),4)
    if a == b:
        print('El seno y el coseno se encuentran en: √2/2 , para este caso en {}'.format(a))
    hx.append(valor)
    hy.append(mt.sin(valor))
    tx.append(valor)
    ty.append(mt.cos(valor))
    #area = acierto / totales
#aca termina la estructura de repeticion
plt.figure(figsize=(5, 5))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.scatter(hx,hy, marker="*", color="pink", label='Sin(x)')
plt.scatter(tx,ty, marker="o", color="cyan", label='Cos(x)')
plt.legend()
plt.show()
```

```python
#Replicar los eventos en numeros base 10, logaritmicos
#Calcular area entre curvas
def SinCos(replicas:int):
    hit, miss = 0, 0
    xhit = []
    yhit = []
    xmiss = []
    ymiss = []
    for i in range(replicas):
        _x = rnd.random()
        _y = rnd.random()
        seno = mt.sin(_x)
        coseno = mt.cos(_x)
        if coseno > seno:
            if _y > seno and _y < coseno:
                hit += 1
                xhit.append(_x)
                yhit.append(_y)
            else:
                xmiss.append(_x)
                ymiss.append(_y)
        else:
            if _y < seno and _y > coseno:
                hit += 1
                xhit.append(_x)
                yhit.append(_y)
            else:
                xmiss.append(_x)
                ymiss.append(_y)
    area = hit/replicas
    return area
```

```python
SinCos(100000)
```

```python
#reportar los valores obtenidos y su comportamiento, lo voy a guardar en un vector
exp = []
for i in tqdm(range(1, 10001), desc="Calculando Seno-Coseno"):
    exp.append(SinCos(i))
promedio = st.mean(exp)
```

```python
plt.figure(figsize=(7, 7))
plt.plot(exp, color='red', linewidth=2, label='Experimentos')
plt.axhline(y=promedio, color='b', linestyle='--', label=f'y=promedio {promedio:,.2f}')
plt.axhline(y=promedio*0.95, color='y', linestyle='--', label=f'y=promedio -5%')
plt.axhline(y=promedio*1.05, color='y', linestyle='--', label=f'y=promedio +5%')
plt.xlabel('Experimentos')
plt.ylabel('Area Aproximada entre sen y cos')
plt.legend()
plt.show()
```
